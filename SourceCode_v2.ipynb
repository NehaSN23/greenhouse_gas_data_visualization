```python
# Define the source path for the files
src_path = "/Users/nehanavarkar/Desktop/Sem3/INFSCI 2415 - INFORMATION VISUALIZATION/IV_Project/dataset_emission/Combine/"

# File names
file_names = [
    'agriculture_emissions_dataset.csv',
    'buildings_emissions_dataset.csv',
    'fluorinated_gases_emissions_dataset.csv',
    'fossil_fuel_operations_emissions_dataset.csv',
    'manufacturing_emissions_dataset.csv',
    'mineral_extraction_emissions_dataset.csv',
    'power_emissions_dataset.csv',
    'waste_emissions_dataset.csv'
]

# Combine the full paths for each file
file_paths = [src_path + file_name for file_name in file_names]

# Initialize list to store DataFrames
dataframes = []

for file_path in file_paths:
    try:
        # Load the file
        df = pd.read_csv(file_path, low_memory=False)
        
        # Standardize `modified_date` if present
        if 'modified_date' in df.columns and 'created_date' in df.columns:
            # Convert to datetime and coerce errors
            df['modified_date'] = pd.to_datetime(df['modified_date'], errors='coerce')
            df['created_date'] = pd.to_datetime(df['created_date'], errors='coerce')
            # Replace invalid `modified_date` with `created_date`
            df['modified_date'] = df['modified_date'].fillna(df['created_date'])
        
        # Append the cleaned DataFrame
        dataframes.append(df)
    except Exception as e:
        print(f"Error processing file {file_path}: {e}")

# Combine all cleaned DataFrames
try:
    combined_df = pd.concat(dataframes, ignore_index=True)
    
    # Define save path for the combined dataset
    save_path = src_path + "greenhouse_gas_emissions_dataset.csv"
    
    # Save the combined dataset
    combined_df.to_csv(save_path, index=False)
    print(f"Combined dataset saved successfully at: {save_path}")
except Exception as e:
    print(f"Error saving the combined dataset: {e}")

# Print a preview of the combined dataset
print(combined_df.head())
```

```python
import pandas as pd
import plotly.graph_objects as go

# File path for the dataset
file_path = "/Users/nehanavarkar/Desktop/Sem3/INFSCI 2415 - INFORMATION VISUALIZATION/IV_Project/dataset_emission/Combine/greenhouse_gas_emissions_dataset.csv"

# Define categories and their subcategories
categories = {
    "Agriculture": [
        "synthetic-fertilizer-application", "cropland-fires", "enteric-fermentation",
        "manure-management", "other-agriculture-soil-emissions", "rice-cultivation"
    ],
    "Buildings": [
        "other-onsite-fuel-usage", "residential-and-commercial-onsite-fuel-usage"
    ],
    "Flourinated Gas": ["fluorinated-gases"],
    "Fossil Fuel Operations": [
        "coal-mining", "oil-and-gas-production-and-transport",
        "oil-and-gas-refining", "other-fossil-fuel-operations", "solid-fuel-transformation"
    ],
    "Manufacturing": [
        "aluminum", "cement", "chemicals", "other-manufacturing", "pulp-and-paper", "steel"
    ],
    "Mineral Extraction": [
        "bauxite-mining", "copper-mining", "iron-mining", "rock-quarrying", "sand-quarrying"
    ],
    "Power": ["electricity-generation", "other-energy-use"],
    "Waste": [
        "biological-treatment-of-solid-waste-&-biogenic", "incineration-and-open-burning-of-waste",
        "solid-waste-disposal", "wastewater-treatment-and-discharge"
    ],
}

try:
    # Load the dataset
    df = pd.read_csv(file_path, low_memory=False)

    # Ensure 'emissions_quantity' is numeric and filter non-null values
    df['emissions_quantity'] = pd.to_numeric(df['emissions_quantity'], errors='coerce')
    df = df[df['emissions_quantity'].notnull()]

    # Map subcategories to main categories
    subcategory_to_category = {sub: cat for cat, sublist in categories.items() for sub in sublist}
    df['main_category'] = df['original_inventory_sector'].map(subcategory_to_category)

    # Drop rows with unmapped categories
    df = df[df['main_category'].notnull()]

    # Aggregate data by main categories and subcategories
    emissions_agg = df.groupby(['main_category', 'original_inventory_sector'])['emissions_quantity'].sum().reset_index()

    # Prepare nodes and links for the Sankey diagram
    main_categories = emissions_agg['main_category'].unique()
    subcategories = emissions_agg['original_inventory_sector'].unique()
    all_nodes = list(main_categories) + list(subcategories)

    # Create node mapping
    node_mapping = {node: i for i, node in enumerate(all_nodes)}

    # Create Sankey links
    links = {
        "source": emissions_agg['main_category'].map(node_mapping),
        "target": emissions_agg['original_inventory_sector'].map(node_mapping),
        "value": emissions_agg['emissions_quantity']
    }

    # Generate Sankey diagram
    fig = go.Figure(go.Sankey(
        node=dict(
            pad=15,
            thickness=10,
            line=dict(color="black", width=0.5),
            label=all_nodes
        ),
        link=dict(
            source=links["source"],
            target=links["target"],
            value=links["value"]
        )
    ))

    # Update layout
    fig.update_layout(
        title_text="Total Emissions by Main Categories and Subcategories",
        font_size=10,
        template="plotly_white"
    )

    # Show the plot
    fig.show()

    # Save the plot as HTML
    output_file = "/Users/nehanavarkar/Desktop/emissions_sankey.html"
    fig.write_html(output_file)
    print(f"Sankey diagram saved as HTML: {output_file}")

except Exception as e:
    print(f"An error occurred: {e}")
```
